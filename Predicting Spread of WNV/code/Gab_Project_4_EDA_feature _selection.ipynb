```python
mapdata = np.loadtxt('../datasets/mapdata_copyright_openstreetmap_contributors.txt')
aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.1)

# An array of years
years = train['year'].unique()

for year in years:
    fig, axs = plt.subplots(ncols=6, figsize=(20,10))
    fig.tight_layout()
    plt.subplots_adjust(top=1.2, wspace=0.025)
    plt.suptitle(f'{year}',size=30, va='baseline')
    
    for i, month in enumerate([5,6,7,8,9,10]):
        
        #plot base maps
        axs[i].imshow(mapdata, 
               cmap=plt.get_cmap('gray'), 
               extent=lon_lat_box, 
               aspect=aspect)
        axs[i].set_title(f'{month}th month')
        axs[i].set_xticks([])
        axs[i].set_yticks([])


        # Trap Locations
        trap_locations = train[['Longitude','Latitude']].drop_duplicates()
        axs[i].scatter(trap_locations['Longitude'],trap_locations['Latitude'], marker='x',c='black')
        
        # Wnv Present in Trap locations
        wnv = train[(train['year'] == year) & (train['month'] == month) & (train['WnvPresent'] == 1)].drop_duplicates()
        axs[i].scatter(wnv['Longitude'],wnv['Latitude'],marker='o',c='yellow')
```

```python
### Annual Mosquitos Count 
fig, ax = plt.subplots(ncols=4, figsize=(20,10), sharey=True)

years = ['2007', '2009', '2011', '2013']

for i in range(4):
    train.set_index('Date').resample('2W').sum()['NumMosquitos'].sort_index().plot(ax=ax[i])
    ax[i].set_xlim(years[i]+'-01', years[i]+'-12-15')
    ax[i].set_xlabel('')

# Global Setting
fig.suptitle('Annual Mosquitos Count in Chicago', x=0.51, y=1.03, fontsize=15)
ax[0].set_ylabel('Number of Mosquitoes', fontsize=12)
plt.tight_layout(pad=0)

plt.show()
```

**Observations**
Mosquitos breed during the Aug-Oct seasons

```python
### Temperature Seasonality on Mosquitoes Count
fig, axs = plt.subplots(nrows=2, figsize=(20,10), sharex=True)

# Temperature resampled at 2-week Interval
train.set_index('Date').resample('2W').mean()['Tavg'].interpolate(method='time').plot(ax=axs[0], 
                                                                   color='orange', 
                                                                   label='Temperature')
ax[0].set_ylabel('Average Temperature, F', fontsize=12)
ax[0].tick_params(axis='x', which='minor', bottom=False)

# Number of Mosquitos
train.set_index('Date').resample('2W').sum()['NumMosquitos'].interpolate(method='time').plot(ax=axs[1],
                                                                                             label='Mosquito Count')
ax[1].set_ylabel('Number of Mosquito', fontsize=12)

# Global Setting
ax[0].set_title('Seasonal Trend of Mosquito Infestation', fontsize=15)
ax[1].set_xlabel('')
timeframe = pd.date_range('2007-06-01', '2013-10-01', freq='2M')
plt.xticks(ticks=timeframe, labels=[x.strftime('%b') for x in timeframe], fontsize=12)

plt.tight_layout(pad=1)
plt.show()
```

**Observation**  
As temperature increases, the number of mosquitos also increase.

## Correlation Matrix for Continuous variables

```python
numerical_features = train.drop(columns=['Date','Species','Address','Block','Street','Trap','AddressNumberAndStreet'])

# Setup the matplotlib figure
plt.figure(figsize=(16,16))

# Compute correlation matrix
df_corr = numerical_features.corr()

# Set the theme
sns.set_theme(style="white")

# Generate a mask for the upper diagonal
mask = np.triu(np.ones_like(df_corr, dtype=bool))

# Adjust mask and dataframe
mask = mask[1: ,:-1]
df_corr = df_corr.iloc[1:,:-1].copy()

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, sep=1, as_cmap=True)

# Plot heatmap
sns.heatmap(df_corr, annot=True, mask=mask, cmap=cmap, square=True, linewidth=0.5, vmax=1, vmin=-1);
```

### Virus Percentage

```python
train['virus_rate'] = train['WnvPresent'] / train['NumMosquitos']
```

### Polynomial Features

```python
# Create 2 dataframes to represent the features and the target
X = numerical_features.drop(columns='WnvPresent')
y = numerical_features['WnvPresent']

# Instantiate polynomial features
pf = PolynomialFeatures(include_bias=False, degree=2)

# Fit & Transform the features on the order of 2
X_pf = pf.fit_transform(X)

# Create dataframe for features
X_pf = pd.DataFrame(X_pf,columns=pf.get_feature_names(X.columns))

# Display the top 10 features correlated with our target
abs(X_pf.corrwith(y).sort_values(ascending=False))[:10]
```

### Dummify Species

```python
train['Pipiens'] = train.apply(lambda row: 1 if 'PIPIENS' in row['Species'] else 0, axis=1)
train['Restuans'] = train.apply(lambda row: 1 if 'RESTUANS' in row['Species'] else 0, axis=1)
```
